# Predicting Flower Types from Iris Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
l=load_iris()
```

```python
dir(l)
```

```python
x=pd.DataFrame(l.data,columns=l.feature_names)
x.head(3)
```

```python
y=pd.DataFrame(l.target)
y.head(3)
```

```python
y=pd.DataFrame(l.target)
y['flower']=y.apply(lambda x:l.target_names[x])
y=y.drop(0,axis='columns')
```

```python
from sklearn.model_selection import train_test_split
x1,x2,y1,y2=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
reg=SVC()
```

```python
reg.fit(x1,y1)
```

```python
reg.predict(x2)
```

```python
reg.score(x2,y2)
```
